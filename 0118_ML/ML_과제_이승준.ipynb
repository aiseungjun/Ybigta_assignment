```python
# 데이터셋 불러오기
# spaceship_tt.csv 데이터프레임을 이용해, "Transported"값을 분류 대상으로 삼는 decision tree classifier 만들기
import numpy as np
import pandas as pd
df = pd.read_csv('spaceship_tt.csv')
```

```python
df.isna().sum()
```

PassengerId : 승객 ID

HomePlanet : 출발 행성(거주지)

CryoSleep : 취침 방식 여부

Cabin : 객실 종류 및 번호 (port : 좌현, starboard : 우현)

Destination : 목적지

Age : 승객의 나이

VIP : 승객의 VIP 서비스 유무

RoomService, FoodCourt, ShoppingMall, Spa, VRDeck : 승객이 해당 서비스에 대해 지불한 금액

Name : 이름

Transported : 도착 여부

```python
# 모델 라이브러리 import
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# base dataframe
df_base=df[['CryoSleep','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Transported']]
```

```python
df_base.dropna(inplace=True)
```

```python
df_base
```

```python
# Model build
dt_clf = DecisionTreeClassifier(random_state=111)

# Train/test split 진행 (7:3)
X_train, X_test, y_train, y_test = train_test_split(df_base.drop('Transported',axis=1),
                                                    df_base['Transported'], # target variable = Transported
                                                    test_size=0.3, random_state=111)
# 모델 훈련
dt_clf = dt_clf.fit(X_train,y_train)
```

```python
X_train
```

```python
# 평가지표 출력
from sklearn.metrics import accuracy_score
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

# 다양한 모델을 통한 평가

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# 랜덤 포레스트, GBM, XGBoost, LightGBM model 별로 평가 수행
# n_estimators를 변경 가능
# 모델별로 추가 parameter 튜닝 가능
rf_reg = RandomForestRegressor(n_estimators=200)
gbm_reg = GradientBoostingRegressor(n_estimators=200)
xgb_reg = XGBRegressor(n_estimators=200)
lgbm_reg = LGBMRegressor(n_estimators=200)
```

```python
model = rf_reg  # random forest


# 모델과 학습/테스트 데이터 셋을 입력하면 성능 평가 수치를 반환하는 함수 정의
from sklearn.metrics import mean_squared_error # regressor를 위한 평가 지표
def get_model_predict(model, X_train, X_test, y_train, y_test, is_expm1=False):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    if is_expm1:
        y_test = np.expm1(y_test)
        pred = np.expm1(pred)

    mse = mean_squared_error(y_test, pred)

    print('###', model.__class__.__name__, '###')
    print('Mean Squared Error:', mse)
```

```python
get_model_predict(model,X_train, X_test, y_train, y_test, is_expm1=True)
```

```python
# 모델과 학습/테스트 데이터 셋을 입력하면 성능 평가 수치를 반환
def get_two_models_predict(model1, model2, X_train, X_test, y_train, y_test, is_expm1=False):
    model1.fit(X_train, y_train)
    model1_pred = model1.predict(X_test)
    model2.fit(X_train, y_train)
    model2_pred = model2.predict(X_test)

    pred = 0.4*model1_pred + 0.6*model2_pred # 각 모델의 반영 가중치 설정

    if is_expm1 :
        y_test = np.expm1(y_test)
        pred = np.expm1(pred)
    mse = mean_squared_error(y_test, pred)

    print('### Two Models ###')
    print('Mean Squared Error:', mse)
```

```python
model1 = rf_reg
model2 = gbm_reg  # GBM Model (Boosting)

get_two_models_predict(model1, model2, X_train, X_test, y_train, y_test, is_expm1=True)
```

### 과제1
비지도 학습의 목적인 Grouping에 대해 자세히 설명해 주세요! Dataset을 Grouping한다는 것이 무슨 의미인지 간단히 서술해 주시고,

세션에서 언급되었던 Grouping을 위한 두 수단에 대해서 예시 모델을 제시하여 설명해 주시면 됩니다! 예시 모델의 종류에 대해서는 구글링을 통해 조사할 수 있겠습니다.

비지도학습이란 지도학습과 달리 training data에 label, 즉 참조할 y값이 없는 데이터로 학습하는 알고리즘 방법이다. 데이터의 라벨이 없으므로 비지도학습은 데이터의 설명 변수들을 축으로 둔 공간에서의 데이터의 위치를 참조하여 데이터들을 "Grouping"하는 학습이라고 정리할 수 있다. 이러한 비지도학습의 대표적인 예시는 특정한 기준으로 더욱 밀접한 관계에 있는 데이터들끼리 군집을 이루도록 하는 Clustering, 데이터 공간의 차원을 줄여(기존 변수를 삭제하고 새로운 변수를 창출하는 방법도 있다.) 설명 변수들의 설명력을 높이고자 하는 Dimension reduction이 있다.

Clustering의 대표적인 모델로는 K-means(density-based clustering), hierarchical(builds a tree of clusters), DBSCAN(density-based), Gaussian Mixture(to fit arbitrarily shaped data)가 있다. Clustering을 수행한 이후 시각화를 통해(dimension reduction이후 시각화하는 것이 나을 수도 있다.) 자신이 원하는 형태로 군집화 되었는지 확인 후, 원하는 형태가 아니라면 다른 clustering의 알고리즘 원리를 참조하여 가장 적절한 clustering 기법을 찾으면 된다. 

Dimension reduction의 대표적인 모델로는 PCA(mapping to a lesser dimensional space in a way that maximizes the variance), LDA와 GDA(linear and nonlinear discriminant analysis that leverages the kernel function)가 있다. 이런식으로 차원의 저주로에서 벗어나 높은 설명력의 변수들을 추출하는 FE과정은 모델의 performance 향상에 큰 도움이 될 수 있다.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('diabetes.csv')
display(df)
```

```python
df_fe = df.dropna()
df_fe.iloc[:, :-1] = df.iloc[:, :-1].apply(lambda x: (x-x.mean()) / x.std(), axis=0)
display(df_fe)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier(random_state=42)

X_train, X_test, y_train, y_test = train_test_split(df_fe.drop('Outcome',axis=1),
                                                    df_fe['Outcome'],
                                                    test_size=0.3,
                                                    random_state=42)

dt_clf = dt_clf.fit(X_train,y_train) # train
pred = dt_clf.predict(X_test) # test
accuracy = accuracy_score(y_test, pred) # eval

print(accuracy)
```

```python
# voting model
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

lr_clf = LogisticRegression(solver='liblinear')
knn_clf = KNeighborsClassifier(n_neighbors=2)
vt_clf = VotingClassifier(estimators=[('LR', lr_clf), ('KNN',knn_clf), ('DT', dt_clf)])

lr_clf.fit(X_train, y_train)
knn_clf.fit(X_train, y_train)
vt_clf.fit(X_train, y_train)

pred = vt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```
