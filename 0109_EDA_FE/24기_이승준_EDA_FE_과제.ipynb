```python
#Q2
miss_df = pd.DataFrame(index=df.columns, columns=["missing_num", "missing_%"])
miss_num = []
miss_p = []
for col in df.columns:
    null_num = df[col].isna().sum()
    if null_num == 0:
        miss_df.drop(col, inplace=True)     
    else:
        miss_num.append(null_num)
        miss_p.append(null_num*100/df.shape[0])
miss_df["missing_num"] = miss_num
miss_df["missing_%"] = miss_p

miss_df = miss_df.sort_values(by="missing_num", ascending=False)

display(miss_df)
```

```python
# Q3
figure, axis = plt.subplots()
axis = sns.barplot(x=miss_df.index, y=miss_df["missing_%"])
axis.set_xlabel('features')
axis.tick_params(axis = 'x', rotation = 90)
```

### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12
for col in df.columns:
    col_null_p = df[col].isnull().sum() / df.shape[0]
    if col_null_p >= 0.25: # column preprocessing
        df.drop(col, axis=1, inplace=True)
    elif col_null_p >= 0.10: # row preprocessing
        df.dropna(subset=col, inplace=True)
    else:
        if df[col].dtypes != np.object_:
            df[col] = df[col].fillna(df[col].median())
display(df)
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1)
# maximum_nights 변수는 일 수로 표현되어 있고 변수 값들을 관찰해 보면 월(30일) 수로 표현하는 것이 더 나은 모델링임을 유추할 수 있다.
df["maximum_nights_mon"] = df["maximum_nights"].apply(lambda x : x//30)
display(df)
```

```python
#Q13-2)
# 전체적인 하우싱의 퀄리티 스코어에 해당하는 review_scores_rating이 점수로 되있어 비교가 힘들다. 간단하게 평균 이상과 이하로 범주화 하였다.(이상:1, 미만:0)
mean_rating = df["review_scores_rating"].mean()
df["upper_than_mean"] = (df["review_scores_rating"] >= mean_rating).astype(int)

display(df)
```

```python
#Q13-3)
# 하우싱 가격에 범주가 없다. 간단하게 평균 이상과 이하로 범주화 하였다.(이상:1, 미만:0)
mean_price = df["price"].mean()
df["upper_price"] = (df["price"] >= mean_rating).astype(int)

display(df)
```
